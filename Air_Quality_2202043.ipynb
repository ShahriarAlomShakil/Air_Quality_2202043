```python
import pandas as pd

dataset = pd.read_csv('/content/AirQualityData.csv')
dataset.head()
```

```python
dataset = dataset.drop('Date', axis=1)
display(dataset.head())
```

```python
dataset.info()
```

```python
dataset = dataset.drop(['Time', 'DayOfWeek', 'Hour'], axis=1)
dataset.info()
```

```python
x = dataset.drop(columns = ['AirQualityIndex'])
y = dataset['AirQualityIndex']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    x,y, random_state=42, test_size= 0.2
)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
lr.score(xtest,ytest)
```

```python
from sklearn.metrics import mean_squared_error as mse
predicted = lr.predict(xtest)
error = mse(ytest, predicted) ** .5
print(error)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

models = [RandomForestRegressor(), DecisionTreeRegressor(), LinearRegression()]

errors = {}

for model in models:
    model.fit(xtrain,ytrain)
    name = type(model).__name__
    score = model.score(xtest,ytest)*100
    errors[name] = score
```
